```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
```

```python
df.head(10)
```

```python
most_views = df.groupby("utm_source").user_id.count().reset_index()
most_views
```

```python
df["is_click"] = df.ad_click_timestamp.apply(lambda x: True if not pd.isna(x) else False)
```

```python
df
```

```python
ad_clicks = df.groupby(["utm_source", "is_click"]).user_id.count().reset_index()
ad_clicks
```

```python
ad_clicks_pivot = ad_clicks.pivot(columns="is_click", index="utm_source", values="user_id").reset_index()
ad_clicks_pivot
```

```python
ad_clicks_pivot["percentage_clicked"] = ad_clicks_pivot[True] / (ad_clicks_pivot[True] + ad_clicks_pivot[False]) * 100
```

```python
ad_clicks_pivot
```

```python
ab_count = df.groupby("experimental_group").user_id.count().reset_index()
ab_count
```

```python
ab_more_clicks = df.groupby(["experimental_group", "is_click"]).user_id.count().reset_index()
ab_more_clicks
```

```python
a_clicks = df[df.experimental_group == "A"]
a_clicks
```

```python
b_clicks = df[df.experimental_group == "B"]
b_clicks
```

```python
a_clicks_by_day_pivot = a_clicks_by_day.pivot(columns="is_click", index="day", values="user_id")
a_clicks_by_day_pivot
```
